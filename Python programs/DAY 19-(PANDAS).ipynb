```python
def addrowcsv(filename, rowdata):
    df=readcsv(filename)
    df.loc[len(df)]=rowdata
    return df
rowdata=['1010101','statement',123,234,445,123,234,456,678,678,789]
addrowcsv('DATA FILES/income.csv',rowdata)
```

```python
def updaterowcsv(filename, rowindex, rowdata):
    df = readcsv(filename)
    df.loc[rowindex] = rowdata
    df.to_csv(filename, index=False)
    return df

rowdata = ['04000US08','Florida',123,4325,4565,9084,83123,423,987,345,765]
updaterowcsv('Data Files/Income.csv', 6, rowdata)

#df = df.drop('Unnamed: 0', axis=1)
#readCSV('Data Files/Income.csv')
```

```python
def deleterowcsv(filename, rowindex, rowdata):
    df = readcsv(filename)
    df=df.drop(rowindex)
    df.to_csv(filename, index=False)
    return df

rowdata = ['04000US08','Florida',123,4325,4565,9084,83123,423,987,345,765]
deleterowcsv('Data Files/Income.csv', 5, rowdata)

#df = df.drop('Unnamed: 0', axis=1)
#readCSV('Data Files/Income.csv')
```
