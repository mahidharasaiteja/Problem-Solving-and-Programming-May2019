```python
import pandas as pd
df = pd.read_csv('DATA FILES/income.csv')

y=df.values[4]
x=df.columns
```

```python
years=[] 
for i in range (2,len(x)):
    years.append(x[i])   
years
income=[]
for i in range(2,len(y)):
    income.append(y[i])
income
p.xlabel('year')
p.ylabel('income')
p.grid(True)
p.title('INCOME vs YEAR graph for CALIFORNIA state')
p.plot(years,income)
p.show()
```
